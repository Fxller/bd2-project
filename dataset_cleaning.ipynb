```python
import pandas as pd
import numpy as np
import random
import string
from flask_bcrypt import Bcrypt
bcrypt = Bcrypt()
import json
```

## Cleaning dataset degli utenti

```python
df_utenti = pd.read_csv("datasets/users_cleaned.csv")
```

```python

df_utenti = df_utenti.drop(columns=["user_watching", "user_completed","user_onhold","user_dropped","user_plantowatch","user_days_spent_watching", "access_rank", "stats_mean_score", "stats_rewatched", "stats_episodes","user_days_spent_watching"])
df_utenti = df_utenti.reset_index(drop=True)
```

```python
df_utenti
```

```python
def genera_password(lunghezza=10):
    caratteri = string.ascii_letters + string.digits
    return ''.join(random.choices(caratteri, k=lunghezza))

df_utenti['password'] = [genera_password() for _ in range(len(df_utenti))]
```

```python
# Salvataggio username e password in chiaro in un file .txt
with open("datasets/credenziali.txt", "w") as f:
    for _, row in df_utenti.iterrows():
        f.write(f"{row['username']}:{row['password']}\n")

print("Credenziali salvate in datasets/credenziali.txt")

# Criptazione delle password con controllo visivo
for i, pw in enumerate(df_utenti['password']):
    hashed = bcrypt.generate_password_hash(pw).decode('utf-8')
    df_utenti.at[i, 'password'] = hashed
    if i % 100 == 0:
        print(f"{i} password criptate...")

print("Criptazione completata.")
```

```python
domini_email = [
    "gmail.com",
    "yahoo.com",
    "outlook.com",
    "hotmail.com",
    "icloud.com",
    "protonmail.com",
    "aol.com",
    "zoho.com",
    "mail.com",
    "gmx.com"
]

df_utenti['email'] = [
    f"{username}@{domini_email[i % len(domini_email)]}"
    for i, username in enumerate(df_utenti['username'])
]
```

## Cleaning dataset animelist

```python
df_animelist = pd.read_csv("datasets/animelists_cleaned.csv")
```

```python
df_animelist
```

```python
df_animelist.drop(columns=["my_start_date", "my_finish_date", "my_status", "my_rewatching", "my_rewatching_ep", "my_last_updated", "my_tags"], inplace=True)
```

```python
df_animelist.head()
```

### Cambio degli id nelle liste

```python
df_anime['anime_id'] = pd.to_numeric(df_anime['anime_id'], errors='coerce')
df_anime['episodes'] = pd.to_numeric(df_anime['episodes'], errors='coerce')

# Merge e salvataggio diretto su df_animelist
df_animelist = df_animelist.merge(
    df_anime[['anime_id', 'episodes']],
    on='anime_id',
    how='left'
)

df_animelist['terminato'] = df_animelist['my_watched_episodes'] >= df_animelist['episodes']

df_animelist['stato'] = df_animelist['terminato'].apply(lambda x: 'completed' if x else 'watching')

df_animelist.drop(columns=["episodes"], inplace=True)
```

```python
df_animelist
```

```python
from bson import ObjectId
# Generazione nuovi _id
df_utenti['_id'] = [ObjectId() for _ in range(len(df_utenti))]
df_anime['_id'] = [ObjectId() for _ in range(len(df_anime))]

# Mappa username → _id utenti
user_map = df_utenti.set_index('username')['_id'].to_dict()

# Mappa anime_id originale → _id degli anime
anime_map = df_anime.set_index('anime_id')['_id'].to_dict()

# Aggiorna df_animelist con i nuovi id
df_animelist['user_id'] = df_animelist['username'].map(user_map)
df_animelist['anime_id'] = df_animelist['anime_id'].map(anime_map)
```

```python
df_utenti.drop(columns=['user_id'], inplace=True)
df_anime.drop(columns=['anime_id'], inplace=True)
```

## Caricamento su MongoDB

```python
from pymongo import MongoClient

def sposta_id_primo(df):
    cols = list(df.columns)
    cols.remove('_id')
    return df[['_id'] + cols]


df_utenti = sposta_id_primo(df_utenti)
df_anime = sposta_id_primo(df_anime)

# Connessione al database
client = MongoClient("mongodb://localhost:27017/")
db_name = "anime_db"

# Droppa il database
client.drop_database(db_name)
db = client[db_name]
print(f"Database '{db_name}' eliminato e ricreato.")

# Inserimento dati dai DataFrame
users_records = df_utenti.to_dict(orient='records')
anime_records = df_anime.to_dict(orient='records')
animelist_records = df_animelist.to_dict(orient='records')

db.users.insert_many(users_records)
print(f"Inseriti {len(users_records)} utenti.")

db.anime.insert_many(anime_records)
print(f"Inseriti {len(anime_records)} anime.")

db.user_anime_list.insert_many(animelist_records)
print(f"Inseriti {len(animelist_records)} record in animelists.")
```
