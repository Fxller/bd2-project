```python
import pandas as pd
import numpy as np
import random
import string
```

## Cleaning dataset degli utenti

```python
df_utenti = pd.read_csv("datasets/users_cleaned.csv")
```

```python

df_utenti = df_utenti.drop(columns=["user_watching", "user_completed","user_onhold","user_dropped","user_plantowatch","user_days_spent_watching", "access_rank", "stats_mean_score", "stats_rewatched", "stats_episodes","user_days_spent_watching"])
df_utenti = df_utenti.reset_index(drop=True)
```

```python
df_utenti
```

```python
def genera_password(lunghezza=10):
    caratteri = string.ascii_letters + string.digits
    return ''.join(random.choices(caratteri, k=lunghezza))

df_utenti['password'] = [genera_password() for _ in range(len(df_utenti))]
```

```python
# Assumiamo di avere una colonna 'username'
domini_email = [
    "gmail.com",
    "yahoo.com",
    "outlook.com",
    "hotmail.com",
    "icloud.com",
    "protonmail.com",
    "aol.com",
    "zoho.com",
    "mail.com",
    "gmx.com"
]

df_utenti['email'] = [
    f"{username}@{domini_email[i % len(domini_email)]}"
    for i, username in enumerate(df_utenti['username'])
]
```

## Cleaning dataset animelist

```python
df_animelist = pd.read_csv("datasets/animelists_cleaned.csv")
```

```python
df_animelist
```

```python
df_animelist.drop(columns=["my_start_date", "my_finish_date", "my_status", "my_rewatching", "my_rewatching_ep", "my_last_updated", "my_tags"], inplace=True)
```

```python
df_animelist.head()
```

### Cambio degli id nelle liste

```python
df_anime['anime_id'] = pd.to_numeric(df_anime['anime_id'], errors='coerce')
df_anime['episodes'] = pd.to_numeric(df_anime['episodes'], errors='coerce')

# Merge e salvataggio diretto su df_animelist
df_animelist = df_animelist.merge(
    df_anime[['anime_id', 'episodes']],
    on='anime_id',
    how='left'
)

df_animelist['terminato'] = df_animelist['my_watched_episodes'] >= df_animelist['episodes']

df_animelist['stato'] = df_animelist['terminato'].apply(lambda x: 'completed' if x else 'watching')

df_animelist.drop(columns=["episodes"], inplace=True)
```

```python
df_animelist
```

```python
df_utenti = df_utenti.rename(columns={'user_id': '_id'})
df_anime = df_anime.rename(columns={'anime_id': '_id'})
```

```python
user_map = df_utenti.set_index('username')['_id'].to_dict()

# --- Aggiunta della colonna 'user_id' ---
df_animelist['user_id'] = df_animelist['username'].map(user_map)

df_animelist.drop(columns=['username'], inplace=True)
print(df_animelist.head())
```
